```python
#Imports
import matplotlib.pyplot as plt
from numpy import exp, linspace
```

```python
#Defining a function to calculate the effective Yukawa potential
def yukawa(l, m, k, a, r):
    return ((l**2) / (2 * (m**2) * (r**2))) - ((k * exp(-a * r)) / r)


#Defining a function to calculate the potential V(r) = -k/r^4
def Veff(l, m, k, r):
    return ((l**2) / (2 * (m**2) * (r**2))) - (k / (r**4))
```

```python
#Constants
l = 1      #angular momentum
m = 1      #mass
k1 = 1     #constant k for Yukawa
a = 1      #dampening constant for Yukawa
k2 = 0.01  #constant k for effective potential


#Setting lists to evaluate the effective potentials
r_list = linspace(0.1, 2, 100)
yukawa_list = []
Veff_list = []


#Loop to evaluate the effective potentials
for r in r_list:
    yukawa_list.append(yukawa(l, m, k1, a, r))
    Veff_list.append(Veff(l, m, k2, r))
```

```python
#Plotting the effective Yukawa potential and the effective potential Veff(r)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 5))
ax1.plot(r_list, yukawa_list, color = 'limegreen')
ax1.set_xlabel('r')
ax1.set_ylabel('$V_{eff}(r)=(l^2/2mr^2)-k\cdot\exp(-ar) / r$')
ax1.set_title('22.b: Effective Yukawa Potential with $l=m=k=a=1$')
ax1.grid(color = 'darkgray', linewidth = 0.5, linestyle = 'dashed')
ax2.plot(r_list, Veff_list, color = 'dodgerblue')
ax2.set_xlabel('r')
ax2.set_ylabel('$V_{eff}(r)=(l^2/2mr^2)-(k/r^4)$')
ax2.set_title('23.b: Effective Potential with $l=m=1, k=0.01$')
ax2.set_xlim(-2.5, 2.5)
ax2.grid(color = 'darkgray', linewidth = 0.5, linestyle = 'dashed')
```
